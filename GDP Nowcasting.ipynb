```python
def combine_dataset(dat1,dat2):
    dat = pd.concat([dat1,dat2],axis =1)
    return dat
```

```python
q.ApiConfig.api_key = 'GeZxMzsDK5j3ec4JZxsm'
dat_dict = {}
# Create a Pandas Excel writer using XlsxWriter as the engine.
writer = pd.ExcelWriter('World Band Data.xlsx', engine='xlsxwriter')

#create new dataset in mongodb
myclient = pymongo.MongoClient("mongodb://localhost:27017/")
mydb = myclient["World_Bank_Data"]
for country in ['USA','TUV']:
    print('Start request',country, 'data')
    for indicator in Indicator.Code.values:
        print('Start request',country, indicator,'data')
        code = 'WWDI/'+ country + '_' + indicator

        try:
            dat1 = q.get(code, start_date="1950-12-31", 
                    end_date="2019-12-31",column_index='1')
            dat1.columns=[indicator]
            if list(Indicator.Code.values).index(indicator)==0:
                dat = dat1
                continue;
            dat = combine_dataset(dat,dat1)
        except:
            print('Jump',indicator)
            continue;
    #insert data to mongodb dataset
    mycol = mydb[country]
    mycol.insert_many(dat.reset_index().to_dict('records'))
    #Write each dataframe to a different worksheet.
    dat.to_excel(writer, sheet_name=country)
    dat_dict[country] = dat
writer.save()
```
