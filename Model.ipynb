```python
import pandas as pd
import numpy as np
```

```python
dat = pd.read_csv('Dat_Merge.csv',index_col = 'Date')
```

```python
dat = dat.drop(columns=['Unnamed: 0'])
```

```python
dat.isnull().sum()
```

```python
np.corrcoef(dat.dropna().T)
```

```python
dat1 = dat.iloc[:230,:]
```

```python
dat2 = dat1.loc['2000-01-01':]
```

```python
drop_list = []
remain_list = []
for name in dat2.columns:
    if dat.loc[:,[name]].isnull().sum().values>100:
        drop_list.append(name)
    else:
        remain_list.append(name)
```

```python
dat3 = dat2.loc[:,remain_list]
```

```python
remain_list
```

```python
var_list=remain_list.copy()
var_list.remove('Rtn')
var_list.remove('Value')
```

```python
dat4  = dat3.fillna(np.mean(dat3))
```

```python
dat4.head()
```

```python
import statsmodels.api as sm # import statsmodels 

X = dat4.loc[:,var_list] ## X usually means our input variables (or independent variables)
y = dat4["Rtn"] ## Y usually means our output/dependent variable
X = sm.add_constant(X) ## let's add an intercept (beta_0) to our model

# Note the difference in argument order
model = sm.OLS(y, X).fit() ## sm.OLS(output, input)
predictions = model.predict(X)

# Print out the statistics
model.summary()
```

```python
from sklearn.feature_selection import RFECV
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
estimator = LinearRegression()
selector = RFECV(estimator, step=1, cv=10)
selector = selector.fit(X, y)
```

```python
selector.ranking_
```

```python
final_list = [var_list[i] for i in range(len(var_list)) if selector.ranking_[i]<=1]
```

```python
X = dat4.loc[:,final_list] ## X usually means our input variables (or independent variables)
y = dat4["Rtn"] ## Y usually means our output/dependent variable
X = sm.add_constant(X) ## let's add an intercept (beta_0) to our model

# Note the difference in argument order
model = sm.OLS(y, X).fit() ## sm.OLS(output, input)
predictions = model.predict(X)

# Print out the statistics
model.summary()
```

```python
np.sum((predictions-dat4.Rtn)**2)/len(predictions)
```

```python
from sklearn.model_selection import train_test_split
X = dat4.loc[:,var_list] 
y = dat4["Rtn"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
#Try different methods
from sklearn.ensemble import AdaBoostRegressor
adbr = AdaBoostRegressor(random_state=0, n_estimators=100)
adbr.fit(X_train, y_train)  
print('Importances:')
print(adbr.feature_importances_)  

pred_ada = adbr.predict(X_test)  
train_ada = adbr.predict(X_train)
```

```python
def mse(y_pred,y):
    return np.sum((y_pred-y)**2)/len(y)
```

```python
print('Train MSE',mse(pred_ada,y_test))
print('Test MSE',mse(train_ada,y_train))
```
