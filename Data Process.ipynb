```python
#define a function to read 47 indices data
def read_data(n):
    data_dict={}
    for i in range(1,n+1):
        if i <10:
            data_dict['M0'+ str(i)] = pd.read_excel(r'C:\Users\Victor\Desktop\Intern\Macro Supprise\Data\M0'+str(i)+'.xlsx',na_values=['- -'])
        else:
            data_dict['M'+ str(i)] = pd.read_excel(r'C:\Users\Victor\Desktop\Intern\Macro Supprise\Data\M'+str(i)+'.xlsx',na_values=['- -'])
    return data_dict
```

```python
#extract only dictionary names
keys_old = list(dat_dict.keys())
```

```python
#rename the keys in the dictionary
for i in range(len(index_name)):
    dat_dict[index_name[i]] = dat_dict.pop(keys_old[i])
```

```python
#extract only useful information
for name in index_name:
    dat_dict[name] = dat_dict[name].loc[:,['Date','Period','Srv Avg','Actual']]
```

```python
#define a fucntion to clean the data in case the data type is string
def clean_dat(dat,column_name):
    
    list1 = dat[column_name].values
    list2 = []
    for i in list1:
        if type(i)==str:
            i = i.replace('k','').replace('b','').replace('$','').replace('m','').replace('- ','-')
            i = float(i)
        list2.append(i) 
   
    return list2
```

```python
#clean the data and caculate surprise
for name in index_name:
    dat_dict[name]['Actual'] = clean_dat(dat_dict[name],'Actual')
    dat_dict[name]['Srv Avg'] = clean_dat(dat_dict[name],'Srv Avg')
    dat_dict[name]['Surprise'] = dat_dict[name]['Actual']-dat_dict[name]['Srv Avg']
```

```python
for name in index_name:
    dat_dict[name]['NormSurp'] = (dat_dict[name].Surprise-np.mean(dat_dict[name].Surprise))/np.std(dat_dict[name].Surprise)
```

```python
from datetime import datetime
from datetime import timedelta
def covert_time(dat_dict,name):
    a = [pd.Timestamp(i) for i in dat_dict[name].Date.values]
    a = [i.to_pydatetime() for i in a]
    Date_cut = [i +timedelta(days=30) for i in a]
    Date_cut = [datetime(i.year,i.month,day=1) for i in Date_cut]
    return Date_cut
```

```python
for name in index_use:
    Date_cut = covert_time(dat_dict_use,name)
    dat_dict_use[name]['Date_cut'] =Date_cut
```

```python
dat_dict_merge = {}
for name in index_use:
    temp_dat= dat_dict_use[name].loc[:,['Date_cut','NormSurp']]
    temp_dat.columns = ['Date',name]
    dat_dict_merge[name] = pd.DataFrame({name:temp_dat[name].values,'Date' : temp_dat['Date'].values})
```

```python
dat_merge = dat_dict_merge['NFP TCH Index'].drop_duplicates(subset='Date', keep='last')
for name in index_use[1:]:
    dat_merge = pd.merge(dat_merge, dat_dict_merge[name].drop_duplicates(subset='Date', keep='last'), 
         on="Date", how = 'outer')
#dat_dict_merge['NFP TCH Index'].drop_duplicates(subset='Date', keep='last')
```
